```python
import numpy as np
import random
```

```python
def get_square(t, j, i):
  return list(t[int(j/3)*3 + 0][int(i/3)*3:int(i/3)*3+3]) + list(t[int(j/3)*3 + 1][int(i/3)*3:int(i/3)*3+3]) + list(t[int(j/3)*3 + 2][int(i/3)*3:int(i/3)*3+3]) 

def get_row(t, v):
  return list(set(t[v]))

def is_solved(t):
  for row in t:
    if 0 in set(row):
      return False
  return True

def solve(puzzle):
  while not is_solved(puzzle):
    for i in range(9):
      for j in range(9):
        if puzzle[i][j] == 0:
          square = get_square(puzzle, i, j)

          temp = puzzle.copy()
          row = get_row(temp, i)

          temp = temp.transpose()
          col = get_row(temp, j)

          remaining = [i for i in range(1, 10)]
          seen = square + row + col

          for v in seen:
            if v in remaining:
              remaining.remove(v)

          if len(remaining) == 1 and puzzle[i][j] == 0:
            puzzle[i][j] = remaining[0]

  return puzzle
```

```python
puzzle_0 = np.array([
                    [5, 3, 0, 0, 7, 0, 0, 0, 0],
                    [6, 0, 0, 1, 9, 5, 0, 0, 0],
                    [0, 9, 8, 0, 0, 0, 0, 6, 0],
                    [8, 0, 0, 0, 6, 0, 0, 0, 3],
                    [4, 0, 0, 8, 0, 3, 0, 0, 1],
                    [7, 0, 0, 0, 2, 0, 0, 0, 6],
                    [0, 6, 0, 0, 0, 0, 2, 8, 0],
                    [0, 0, 0, 4, 1, 9, 0, 0, 5],
                    [0, 0, 0, 0, 8, 0, 0, 7, 9]
                   ])

solve(puzzle_0)
```
